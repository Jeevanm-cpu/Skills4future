```python
# load the dataset -sns
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
data =load_iris()
data
```

```python
x = data.data
y = data.target
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtr,xte,ytr,yte = train_test_split(x,y,test_size=0.2,random_state=16)
```

```python
xtr.shape
```

```python
ytr.shape
```

```python
# standardization -scaling
from sklearn.preprocessing import StandardScaler
ss =StandardScaler()
```

```python
xtr_scaled = ss.fit_transform(xtr)
xtr_scaled
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
# instantiate the LR class
lr_model = LogisticRegression()
lr_model.fit(xtr_scaled,ytr)
```

```python
# prediction
lr_pred = lr_model.predict(xte)
```

```python
lr_pred
```

```python
# evaluation
acc = accuracy_score(yte,lr_pred)
print(f'The accuracy score is {acc*100:.2f}%')
```

```python
# confusion matrix
cm = confusion_matrix(yte,lr_pred)
cm
```

```python
# classification report
cr = classification_report(yte,lr_pred)
print(cr)
```

<h1>Decision Trees</h1>

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv(r'C:\Users\Lenovo\OneDrive\Documents\Renewable_Energy_Adoption.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
# feature and target selection
x =df[['carbon_emissions', 'energy_output', 'renewability_index','cost_efficiency']]
y =df['adoption']
```

```python
# train test split 
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
model = DecisionTreeClassifier(max_depth=3,random_state=42)
```

```python
model.fit(xtrain,ytrain) #train the model
```

```python
# predict
y_pred = model.predict(xtest)
y_pred
```

```python
#evaluate the model
acc=accuracy_score(ytest,y_pred)
print(f'The accuracy score for DTC is {acc*100:.2f}%')
```

```python
cm = confusion_matrix(ytest,y_pred)
sns.heatmap(cm,annot=True,cmap='Blues',xticklabels=['Non Adoption','Adoption'],yticklabels=['Non Adoption','Adoption'])
plt.title('Heatmap of confusion Matrix')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,10))
plot_tree(model,filled=True,feature_names=x.columns,class_names=['Non Adoption','Adoption'])
plt.show()
```

```python
cr =classification_report(ytest,y_pred)
print(cr)
```
